Desafio 002

Crie um script Python que leia o dia, mês e o ano de nascimento de uma pessoa e mostra uma mensagem com a data formatada.

```python
dia = int(input('Digite o dia do seu nascimento: '))
mes = int(input('Digite o mês do seu nascimento: '))
ano = int(input('Digite o ano do seu nascimento: '))
print('Sua data de nascimento é: {}/{}/{}'.format(dia, mes, ano))
```

**Desafio 003**

Crie um programa que leia dois números e mostra a soma entre eles

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))
print('A soma entre {} e {} é {}'.format(n1, n2, n1 + n2))
```

**Desafio 004**

Faça um programa que leia algo pelo teclado e mostra na tela seu tipo primitivo e todas as informações possíveis sobre ela

```python
v = input('Digite algo: ')
print('Seu tipo primitivo é {}'.format(type(v)))
print('É numérico: {}'.format(v.isnumeric()))
print('É alfanúmerico: {}'.format(v.isalnum()))
print('É alfabético: {}'.format(v.isalpha()))
print('Está em maiúsculo: {}'.format(v.isupper()))
```
