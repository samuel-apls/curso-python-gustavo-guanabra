**ex004**

Faça um programa que leia algo pelo teclado e mostre na tela o seu tipo primitivo e todas as informações possíveis sobre elas.

```python
valor = input(print('Digite algo: '))
print("O tipo primitivo desse valor é: {}".format(type(valor)))
print('Só tem espaços? {}'.format(valor.isspace()))
print('É um número? {}'.format(valor.isnumeric()))
print('É alfabético? {}'.format(valor.isalpha()))
print('É alfanumérico? {}'.format(valor.isalnum()))
print('Está em maiúsculo? {}'.format(valor.isupper()))
print('Está em minúsculo? {}'.format(valor.islower()))
print('Está em capitalizado? {}'.format(valor.istitle()))
```
